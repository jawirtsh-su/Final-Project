```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing vodka infomation",
vodka_url="http://www.efficientdrinker.com/liquor/"
vodka_data=pd.read_html(vodka_url)
vodka_df=vodka_data[0]
vodka_df=vodka_df.rename(columns={'Brand or Type':'Brand'})
vodka_df['Type']='Liquor'
vodka_df=vodka_df.drop(columns=['ABV','Efficiency','Carbs (g)'])
vodka_df.sample(3)

#importing beer infomation
beer_url="http://www.efficientdrinker.com/beer/complete-calories-abv/"
beer_data=pd.read_html(beer_url)
beer_df=beer_data[0]
beer_df['Type']='Beer'
beer_df=beer_df.drop(columns=['ABV','Efficiency'])
beer_df.head(5)

#importing malt infomation
malt_url="http://www.efficientdrinker.com/wine/malternatives/"
malt_data=pd.read_html(malt_url)
malt_df=malt_data[0]
malt_df['Type']='Malt'
malt_df=malt_df.drop(columns=['ABV','Efficiency'])
malt_df.head(5)
```

```python
alc_df=pd.concat([malt_df,beer_df,vodka_df]) #joining all three of the dataframes together\n",
alc_df=alc_df.reset_index() #reseting the index for the one dataframe\n",

#changing the values in data frame to lowercase,
alc_df["Brand"]=alc_df["Brand"].str.lower()
alc_df["Type"]=alc_df["Type"].str.lower()

alc_df.sample(10)
```

```python
def lower_cal(kind,name):
    indi=alc_df[alc_df['Brand']==name].index.item()
    num_cals=alc_df.get_value(indi,'Calories')

    suggest_lowcal=alc_df[ (alc_df["Type"]==kind) & (alc_df["Calories"]<num_cals)]
    suggestions=suggest_lowcal.sample(5)
    
    return suggestions

lower_cal("liquor","fireball")

def get_cal(kind,name): #function that returns the amount of calories in one serving
    indi=alc_df[alc_df['Brand']==name].index.item()
    num_cals=alc_df.get_value(indi,'Calories')
    
    return num_cals
```

```python
try:
    
    beer_total=0
    vodka_total=0
    malt_total=0
    print("Alcohol Calorie Counter \n")
    
    while True:
        
        user_type=input("Please enter a type of alcohol. Your choices are Beer, liquor, and Malt. Enter STOP if your finished. \n")
        
        #test to see if user enters stop while ignoring case
        if(user_type.casefold()=='STOP'.casefold()):
            cal_total=beer_total + vodka_total + malt_total
            print("On your night out you drank a total of %.2f calories" %(cal_total ))
            break
            
        #test to see if user enters beer while ignoring case
        elif(user_type.casefold()=="Beer".casefold()):
            user_type="beer"
            while True:
                user_brand=input("Please enter a brand of Beer you had. Enter STOP if your done.\n")
                user_brand=user_brand.lower()

                if(user_brand=="stop"):
                    break

                #preforms calculation and adds calorie count to running sum
                num_drinks=int(input("Please enter the number of drinks of %s you had.\n" %(user_brand)))
                cal_count_beer=get_cal(user_type,user_brand)
                cal_count_beer=cal_count_beer*num_drinks
                beer_total=beer_total+cal_count_beer
                print("While having %d drinks of %s you consumed %.2f calories" %(num_drinks,user_brand,cal_count_beer))
                print("Here are some lower calorie alternatives")
                display(lower_cal(user_type,user_brand))
                
         #test to see if user enters vodka while ignoring case
        elif(user_type.casefold()=="Liquor".casefold()):
            user_type="liquor"
            while True:
                user_brand=input("Please enter a brand of liquor you had. Enter STOP if your done.\n")
                user_brand=user_brand.lower()

                if(user_brand=="stop"):
                    break
                    
                #preforms calculation and adds calorie count to running sum
                num_drinks=int(input("Please enter the number of shots of %s you had.\n" %(user_brand)))
                cal_count_vodka=get_cal(user_type,user_brand)
                cal_count_vodka=cal_count_vodka*num_drinks
                vodka_total=vodka_total+cal_count_vodka
                print("While having %d shots of %s you consumed %.2f calories" %(num_drinks,user_brand,cal_count_vodka))
                print("Here are some lower calorie alternatives")
                display(lower_cal(user_type,user_brand))
                
        #test to see if user enters beer while ignoring case
        elif(user_type.casefold()=="Malt".casefold()):
            user_type="malt"
            while True:
                user_brand=input("Please enter a brand of Malt liquor you had. Enter STOP if when your done.\n")
                user_brand=user_brand.lower()

                if(user_brand=="stop"):
                    break
                    
                #preforms calculation and adds calorie count to running sum
                num_drinks=int(input("Please enter the number of drinks of %s you had.\n" %(user_brand)))
                cal_count_malt=get_cal(user_type,user_brand)
                cal_count_malt=cal_count_malt*num_drinks
                malt_total=malt_total+cal_count_malt
                print("While having %d drinks of %s you consumed %.2f calories" %(num_drinks,user_brand,cal_count_malt))
                print("Here are some lower calorie alternatives")
                display(lower_cal(user_type,user_brand))

         #check for the user       
        else:
            print("Im sorry one of our alcohol types isn't %s" %(user_type)) 
#prevents error message 
except ValueError:
    print("ERROR: Either that brand was not located in our database or you didn't enter a proper number of drinks")
```
